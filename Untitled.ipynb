```python
# sandbox
from PIL import Image
import matplotlib.pyplot as plt

def crop_stuff(im):
    c=im
    w,h=c.size
    dark,light = c.getextrema()
    upperlim = light * h
    left_w=0
    right_w = w
    up_h=0
    down_h=h
    for i in range(w):
        if sum([im.getpixel((i,j)) for j in range(h)]) > upperlim-10:
            left_w=i
        else:
            break
    for i in range(w-1,0,-1):
        if sum([im.getpixel((i,j)) for j in range(h)]) > upperlim-10:
            right_w=i
        else:
            break
    upperlim = light*w
    for j in range(h):
        if sum([im.getpixel((i,j)) for i in range(w)]) > upperlim-10:
            up_h = j
        else:
            break
    for j in range(h-1,0,-1):
        if sum([im.getpixel((i,j)) for i in range(w)]) > upperlim-10:
            down_h = j
        else:
            break
    if left_w>10:
        left_w = left_w-10
    else:
        left_w = 0
    if right_w+10<=w:
        right_w = right_w+10
    else:
        right_w = w
    if up_h>10:
        up_h = up_h-10
    else:
        up_h = 0
    if down_h + 10 <= h:
        down_h = down_h + 10
    else:
        down_h = h
    return c.crop((left_w,up_h,right_w,down_h))
```

```python
def preprocess(im,pageno,default_linenum):
    # for each line:
     # 1. count range of pixels present
     # 2. if pixel range too short(<1 letter, maybe h//20), just continue on.
     # 3. if too few pixels, continue
     # 4. if the rate changes from "unaccepted" to "over some threshold" we will mark this.
     # 5. we need to prune furigana: do the crop((entry,0,entry+10,h)) and crop((entry-5,0,entry+10,h))
    print(f"page {pageno}")
    c = crop_stuff(im)
    try:
        dark,light = c.getextrema()
    except TypeError:
        return None
    w,h = c.size
    upperlim = light * h
    y = [sum([c.getpixel((i,j)) for j in range(h)]) for i in range(w)]
    prev_res = False
    filtered = []
    final = []
    rates = []
    # ----- find where to ~~`.
    try:
        for i in range(w):
            up,down = 0,h-1
            while up < h and c.getpixel((i,up)) >= light-1:
                up += 1
            while down > 0 and c.getpixel((i,down)) >= light-1:
                down -= 1
            rate = (upperlim - y[i])/(down+1-up)
            rates.append(rate)
            if up + 20 <= down and (prev_res is False) and rate > 1:
                filtered.append(i)
                prev_res = True
            elif rate < 0.1:
                prev_res = False
    except ZeroDivisionError:
        print(up,down)
        print(len(rates))
    for entry in filtered:
        one = crop_stuff(c.crop((entry,0,entry+10,h)))
        two = crop_stuff(c.crop((entry-5,0,entry+10,h)))
        if abs(one.size[1]-two.size[1])<20:
            final.append(entry)
    # ----- line number & linewidth evaluation
    linenum_flag=False
    try:
        for mae,ushiro in zip(final,final[1:]):
            if abs(ushiro-mae-final[1]-final[0]) > 50:
                linenum_flag = True
        if linenum_flag is True:
            linenum = default_linenum
        elif len(final)>1:
            linenum = len(final)
        else:
            linenum = None
    except IndexError:
        print(filtered)
        print("this image isn't text.")
        return None
    try:
        linewidth = 0.68*(final[-1]-final[0])/linenum
    except (TypeError,IndexError):
        print(filtered)
        print("this image isn't text.")
        return None
    # ----- filter
    should_we_filter = [True for i in range(w)]
    for entry in final:
        for i in range(int(linewidth)):
            try:
                should_we_filter[entry+i-2] = False
                # should_we_filter[entry-2] = 2
            except IndexError:
                break
    for ind, yes_or_no in enumerate(should_we_filter):
        if yes_or_no is True:
            c.paste(light,box=((ind,0,ind+1,h)))
        elif yes_or_no is 2:
            c.paste(dark,box=((ind,0,ind+1,h)))
    #c.save(f"./processed/bake{pageno:04d}_s.png")
    #plt.plot(range(w),y,"b--")
    #plt.show()
    #print(junctions)
    #print(filtered)
    #print([i+int(linewidth) for i in filtered])
    #print(len(filtered))
    return c
```

```python
def split_into_halves(im,pageno):
    w,h = im.size
    im = im.crop((w//20,h//11,19*w//20,h))
    w,h = im.size
    im1 = preprocess(im.crop((0,0,w,h//2)),pageno,18)
    if im1 is not None:
        im1.save(f"./processed/bake{pageno:04d}_1.png")
    im2 = preprocess(im.crop((0,h//2,w,h)),pageno,18)
    if im2 is not None:
        im2.save(f"./processed/bake{pageno:04d}_2.png")

def run_this(pageno):
    try:
        im = Image.open(f"bakemono{pageno:04d}_s.png")
        split_into_halves(im,pageno)
    except FileNotFoundError:
        im = Image.open(f"bakemono{pageno:04d}_s.jpg")
        # im.save(f"./processed/bakemono{pageno:04d}_s.jpg")

for i in range(10,444):
    run_this(i)
```
