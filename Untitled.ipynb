```python
# sandbox
from PIL import Image
import matplotlib.pyplot as plt

def crop_stuff(im):
    c=im
    w,h=c.size
    dark,light = c.getextrema()
    upperlim = light * h
    left_w=0
    right_w = w
    up_h=0
    down_h=h
    for i in range(w):
        if sum([im.getpixel((i,j)) for j in range(h)]) > upperlim-10:
            left_w=i
        else:
            break
    for i in range(w-1,0,-1):
        if sum([im.getpixel((i,j)) for j in range(h)]) > upperlim-10:
            right_w=i
        else:
            break
    upperlim = light*w
    for j in range(h):
        if sum([im.getpixel((i,j)) for i in range(w)]) > upperlim-10:
            up_h = j
        else:
            break
    for j in range(h-1,0,-1):
        if sum([im.getpixel((i,j)) for i in range(w)]) > upperlim-10:
            down_h = j
        else:
            break
    if left_w>10:
        left_w = left_w-10
    else:
        left_w = 0
    if right_w+10<=w:
        right_w = right_w+10
    else:
        right_w = w
    if up_h>10:
        up_h = up_h-10
    else:
        up_h = 0
    if down_h + 10 <= h:
        down_h = down_h + 10
    else:
        down_h = h
    return c.crop((left_w,up_h,right_w,down_h))
```

```python
def preprocess(im,pageno):
    # instead of doing this bullshit, maybe we can go line by line?
    print(f"page {pageno}")
    c = crop_stuff(im)
    try:
        dark,light = c.getextrema()
    except TypeError:
        return None
    w,h = c.size
    y = [sum([c.getpixel((i,j)) for j in range(h)]) for i in range(w)]
    upperlim = light * h
    junctions = [i for i in range(w-1) 
                     if y[i]>upperlim*0.97 and y[i+1]<upperlim*0.97]
    filtered = [place for ind,place in enumerate(junctions) if 
                    ind==0 or junctions[ind] - junctions[ind-1] > 5 ]
    final=[]
    for entry in filtered:
        one = crop_stuff(c.crop((entry,0,entry+10,h)))
        two = crop_stuff(c.crop((entry-5,0,entry+10,h)))
        if abs(one.size[1]-two.size[1])<20:
            final.append(entry)
    final=filtered
    try:
        gap = (final[-1]-final[0])/(len(final)-1)
    except (ZeroDivisionError, IndexError):
        # im.save(f"./processed/bakemono{pageno:04d}_s.png")
        print("this image isn't text.")
        return None
    linewidth = gap*0.66
    should_we_filter = [True for i in range(w)]
    for entry in final:
        for i in range(int(linewidth)):
            try:
                should_we_filter[entry+i-2] = False
                should_we_filter[entry-2] = 2
            except IndexError:
                break
    for ind, yes_or_no in enumerate(should_we_filter):
        if yes_or_no is True:
            c.paste(light,box=((ind,0,ind+1,h)))
        elif yes_or_no is 2:
            c.paste(dark,box=((ind,0,ind+1,h)))
    #c.save(f"./processed/bake{pageno:04d}_s.png")
    #plt.plot(range(w),y,"b--")
    #plt.show()
    #print(junctions)
    #print(filtered)
    #print([i+int(linewidth) for i in filtered])
    #print(len(filtered))
    return c
```

```python
def split_into_halves(im,pageno):
    w,h = im.size
    im = im.crop((w//20,h//11,19*w//20,h))
    w,h = im.size
    im1 = preprocess(im.crop((0,0,w,h//2)),pageno)
    if im1 is not None:
        im1.save(f"./processed/bake{pageno:04d}_1_comp.png")
    im2 = preprocess(im.crop((0,h//2,w,h)),pageno)
    if im2 is not None:
        im2.save(f"./processed/bake{pageno:04d}_2_comp.png")

def run_this(pageno):
    try:
        im = Image.open(f"bakemono{pageno:04d}_s.png")
        split_into_halves(im,pageno)
    except FileNotFoundError:
        im = Image.open(f"bakemono{pageno:04d}_s.jpg")
        # im.save(f"./processed/bakemono{pageno:04d}_s.jpg")
```

```python
from PIL import Image

for pageno in range(10,446):
    try:
        im = Image.open(f"bakemono{pageno:04d}_s.png")
    except FileNotFoundError:
        im = Image.open(f"bakemono{pageno:04d}_s.jpg")
        im.save(f"./processed/page{pageno:03d}mod.png")
        
    (w,h) = im.size
    c = im.crop((0,70,w,h))
    (w,h) = c.size

    odd = [86+int(36.8*i)-2 for i in range(18)]
    even = [84+int(36.9*i)-2 for i in range(18)]
    colorwhite = [True for i in range(w)]
    pattern = []
    textwidth = 0

    if pageno%2==0:
        pattern = even
        textwidth = 22
    else:
        pattern = odd
        textwidth = 22

    for i in pattern:
        for j in range(textwidth):
            colorwhite[i+j] = False

    for i in range(w): #has length of w
        if colorwhite[i] is True:
            for j in range(h):
                c.putpixel((i,j),15)

    c.save(f"./processed/page{pageno:03d}mod.png")
        
            
```

```python
from PIL import Image

pageno = 13
im = Image.open(f"bakemono{pageno:04d}_s.png")
(w,h) = im.size
c = im.crop((0,100,w,h))
(w,h) = c.size
```

```python
sums = []

for i in range(w):
    sums.append(0)
    for j in range(w):
        sums[i] += c.getpixel((i,j))
```

```python
import matplotlib.pyplot as plt

the_plot = plt.plot(range(w),sums,"b--")
plt.show()
```

```python
inds = []
for i in range(w):
    if sums[i] > 11500:
        inds.append(1)
    else:
        inds.append(0)
plt.plot(range(len(inds)),inds,"r--")
plt.show()
```
